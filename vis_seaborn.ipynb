## Seaborn
Refer - https://seaborn.pydata.org/tutorial/function_overview.html

In the seaborn library, the plot that we create is divided into the following various categories:

+ ##### Distribution plots:
This type of plot is used for examining both types of distributions, i.e., univariate and bivariate distribution.
+ ##### Relational plots:
This type of plot is used to understand the relation between the two given variables.
+ ##### Categorical plots:
The categorical plots are used to deals with categories of variables and how we can visualize them

## a. Linechart

```python
sns.relplot(data=mpg, x='model_year',y='mpg',
            kind='line',
            ci=False,
           )
plt.show()
```

```python
sns.relplot(data=mpg, x='model_year',y='mpg',
            kind='line',
            ci=False,
            style='origin',
            hue='origin'
           )
plt.show()
```

### Scatter Plot

```python
pal= sns.set_palette('viridis')
sns.relplot(data=tips, x='total_bill',y='tip',
            kind='scatter',
            size='sex',
            hue='time',
            palette=pal,
            height=3,
            aspect=1.5
           )
plt.show()
```

### Countplot

```python
### Create a count plot of survey responses
cp = sns.countplot(data=tips, x="size")
plt.show()
```

```python
### Create a count plot of survey responses
cp = sns.countplot(data=tips, x="size")

# Show the counts on top of the bars
for bars in cp.containers:
    cp.bar_label(bars)

plt.show()
```
